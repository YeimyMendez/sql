```python
# importamos librerias

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm
import scipy.stats as stats
```

```python
dataf = pd.read_csv('./data/Nacidos_Vivos_en_2016_de_Guadalajara_de_Buga.csv')
```

```python
dataf
```

```python
# medidas de tendencia central
media_p = dataf["Edad del Padre"].mean()
media_m = dataf["Edad de la Madre"].mean()

mediana_p = dataf["Edad del Padre"].median()
mediana_m = dataf["Edad de la Madre"].median()

moda_p = dataf["Edad del Padre"].mode() [0]
moda_m = dataf["Edad de la Madre"].mode() [0]
```

```python
# medidas de dispersion
rango_p = dataf["Edad del Padre"].max() - dataf["Edad del Padre"].min()
rango_m = dataf["Edad de la Madre"].max() - dataf["Edad de la Madre"].min()

varianza_p = dataf["Edad del Padre"].var()
varianza_m = dataf["Edad de la Madre"].var()

desviacions_p = dataf["Edad del Padre"].std()
desviacions_m = dataf["Edad de la Madre"].std()
```

```python
# percentil 75 25
percentil_25_p = dataf["Edad del Padre"].quantile(0.25)
percentil_75_p = dataf["Edad del Padre"].quantile(0.75)

percentil_25_m = dataf["Edad de la Madre"].quantile(0.25)
percentil_75_m = dataf["Edad de la Madre"].quantile(0.75)
```

```python
# calcular IQR
iqr_p = percentil_75_p - percentil_25_p 
iqr_m = percentil_75_m - percentil_25_m 
```

```python
# calcular desviacion absoluta
desviacion_abs_p = (dataf["Edad del Padre"] - media_p).abs() .mean()
desviacion_abs_m = (dataf["Edad de la Madre"] - media_m).abs() .mean()
```

```python
# imprimir resultados
print("Medidas de tendencia central de los padres")
print("-----------------------------------------------")
print(f"Media papá: {media_p:.2f}")
print(f"Media mamá: {media_m:.2f}")

print(f"\nMediana papá: {mediana_p:.2f}")
print(f"Mediana mamá: {mediana_m:.2f}")

print(f"\nModa papá: {moda_p}")
print(f"Moda mamá: {moda_m}")

# medidas de dispersion
print("\nMedidas de dispersion de los padres")
print("-----------------------------------------------")
print(f"Rango papá: {rango_p}")
print(f"Rango mamá: {rango_m}")

print(f"\nVarianza papá: {varianza_p:.2f}")
print(f"Varianza mamá: {varianza_m:.2f}")

print(f"\nDesviacion Standar papá: {desviacions_p:.2f}")
print(f"Desviacion Standar mamá: {desviacions_m:.2f}")

# percentiles
print(f"\npercentil 75 papá: {percentil_75_p:.2f}")
print(f"percentil 25 papá: {percentil_25_p:.2f}")

print(f"\npercentil 75 mamá: {percentil_75_m:.2f}")
print(f"percentil 25 mamá: {percentil_25_m:.2f}")

# IQR
print(f"\nIQR papá: {iqr_p:.2f}")
print(f"IQR Mamá: {iqr_m:.2f}")

# Desviacion absoluta
print(f"\nDesviacion absoluta papá: {desviacion_abs_p:.2f}")
print(f"Desviacion absoluta Mamá: {desviacion_abs_m:.2f}")



```
